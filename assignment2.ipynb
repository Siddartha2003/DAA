```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
```

```python
nltk.download('punkt')
```

```python
text = '''The cats are playing with a ball. There were several cats, and they are all playing. <br>
The word "play" can have different meanings in different contexts.'''
```

```python
def remove_ambiguity(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text
```

```python
def tokenize_text(text):
    sentences = sent_tokenize(text)
    words = [word_tokenize(sentence) for sentence in sentences]
    return sentences, words
```

```python
def stem_words(words):
    ps = PorterStemmer()
    stemmed_words = [[ps.stem(word) for word in sentence] for sentence in words]
    return stemmed_words
```

```python
cleaned_text = remove_ambiguity(text)
print("Cleaned Text:", cleaned_text)
```

```python
sentences, tokenized_words = tokenize_text(cleaned_text)
print("Tokenized Sentences:", sentences)
print("Tokenized Words:", tokenized_words)
```

```python
stemmed_words = stem_words(tokenized_words)
print("Stemmed Words:", stemmed_words)
```
